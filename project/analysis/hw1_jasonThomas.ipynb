2. Add columns to the dataframe containing:
  * The total (East + West) bicycle count (1pt)
  * The hour of the day (1pt)
  * The year (1pt)

```python
#the total column already exists! But I'll make a new one anyway.
df['total_eastWest_bikeCount'] = df['Fremont Bridge East Sidewalk'] + df['Fremont Bridge West Sidewalk']
df.head()
```

```python
from datetime import datetime

df['Date_datetime'] =  pd.to_datetime(df['Date'])
df['hour'] = df['Date_datetime'].dt.hour
df['year'] = df['Date_datetime'].dt.year
```
